```python
import numpy as np

from utils import posterior_samples, plot_function_densities, relative_increase, plot_sample_densities
```

```python
impressions_A = 3000
conversions_A = 175
```

```python
impressions_B = 2500
conversions_B = 150
```

```python
posterior_A, samples_A = posterior_samples(impressions_A, conversions_A)
samples_A.shape
```

```python
posterior_B, samples_B = posterior_samples(impressions_B, conversions_B)
samples_B.shape
```

```python
plot_function_densities(
    [posterior_A, posterior_B],
    ['aposterior density A', 'aposterior density B'],
    x = np.linspace(0, 0.1, 100)
)
```

```python
(samples_B > samples_A).mean()
```

```python
relative_posterior_samples = relative_increase(samples_B, samples_A)
relative_posterior_samples.shape
```

```python
plot_sample_densities([relative_posterior_samples], ['relative posterior sample'])
```
