```python
import numpy as np

from utils import multinomial_posterior_samples, plot_sample_densities, relative_increase
```

```python
costs = np.array([2000, 1000, 500, 0])
```

# A

```python
N_A = 1500
N1_A = 15
N2_A = 43
N3_A = 157
N4_A = N_A - (N1_A + N2_A + N3_A)
N_list_A = [N1_A, N2_A, N3_A, N4_A]
N_list_A
```

```python
posterior_samples_A = multinomial_posterior_samples(N_list_A)
posterior_samples_A.shape
```

```python
plot_sample_densities(posterior_samples_A.T, ['p_{}'.format(c) for c in costs], bins=50)
```

```python
posterior_expected_cost_A = np.dot(posterior_samples_A, costs)
posterior_expected_cost_A.shape
```

```python
plot_sample_densities([posterior_expected_cost_A], ['posterior cost A'], bins=50)
```

# B

```python
N_B = 2000
N1_B = 25
N2_B = 63
N3_B = 252
N4_B = N_B - (N1_B + N2_B + N3_B)
N_list_B = [N1_B, N2_B, N3_B, N4_B]
N_list_B
```

```python
posterior_samples_B = multinomial_posterior_samples(N_list_B)
posterior_samples_B.shape
```

```python
plot_sample_densities(posterior_samples_B.T, ['p_{}'.format(c) for c in costs], bins=50)
```

```python
posterior_expected_cost_B = np.dot(posterior_samples_B, costs)
posterior_expected_cost_B.shape
```

```python
plot_sample_densities([posterior_expected_cost_B], ['posterior cost B'], bins=50)
```

# AB

```python
plot_sample_densities([posterior_expected_cost_A, posterior_expected_cost_B], ['posterior cost A', 'posterior cost B'], bins=50)
```

```python
plot_sample_densities([posterior_expected_cost_B - posterior_expected_cost_A], ['posterior_cost_difference'], bins=50)
```

```python
(posterior_expected_cost_B - posterior_expected_cost_A > 0.5).mean()
```

```python
relative_posterior_samples = relative_increase(posterior_expected_cost_B, posterior_expected_cost_A)
relative_posterior_samples.shape
```

```python
plot_sample_densities([relative_posterior_samples], ['relative posterior sample'])
```
